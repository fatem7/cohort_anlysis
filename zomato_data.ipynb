```python
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Zomato data .csv')
```

```python
df.head()
```

```python
df.info()
```

```python
def handelrate(value):
  value=str(value).split('/')
  value=value[0]
  return float(value)

df['rate']=df['rate'].apply(handelrate)
```

```python
df.head()
```

```python
df.info()
```

```python
sns.countplot(x=df['listed_in(type)']) #makes plot based on the count of this column values
plt.xlabel('Type of resturant')
```

Conclusion: The majority of the restaurants fall into the dining category.

```python
grouped_data=df.groupby('listed_in(type)')['votes'].sum()
grouped_data
```

```python
xlabel=grouped_data.index
ylabel=grouped_data.values
result = pd.DataFrame({'votes':grouped_data})
result
```

```python
plt.plot(xlabel,ylabel,color='green',marker='o')
plt.xlabel('type of resturants',color='orange',size=12)
plt.ylabel('votes',color='orange',size=12)
```
